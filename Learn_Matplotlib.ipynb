**Plotting**
ploting merupakan menarik garis dari titik ke titik 
fungsi dari plot() adalah adalah menggambar line dari titik ke titik.
dengan paramater titik tersebut adalah (x,y)

misal kita ingin menggambar line dari (1,7) ke (7,14)
sehingga , kita memiliki 2 array , yaitu x[1,7] dan y[7,14]



```python
#import library
import matplotlib.pyplot as plt
import numpy as np

#assgin x,y
x = np.array([1,7])
y = np.array([7,14])

#show line
plt.plot(x,y)
plt.show()
```

**Custom markers**

Dapat mengcustom mark dengan menentukan mark nya apa , line , dan warnanya 

line syntax: 

'-' = solid line

':' = dot line 

'--' = dash line

'-.' = solid and dot line

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,7,2,6,2,8,4])


plt.plot(y,'o:r') #berarti mark o dengan garis : dan warna red
plt.show()
```

**Marker Size**

menggunakan keyword markersize atau dapat disingkat dengan ms

**Marker Colour**

marker colour dibedakan menjadi 2 macam ;
 

1.   MarkerEdgeColour
       marker ini merupakan marker dengan colournya hanya bagian luar saja. 
2.   MarkerFaceColour
       marker ini dengan colournya hanya  pada isi atau dalamnya. dan bagian luar dari mark tetap berwarna default





Marker Edge Colour


Untuk menggunakan marker ini gunakan keyword 'markeredgecolour' disingkat menjadi 'mec'

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,7,2,6,2,8,4])


plt.plot(y,marker = 'o',ms=20,mec='r') 
plt.show()
```

Marker Face Colour

Untuk menggunakan markerk ini dapat menggunakan keyword 'markerfacecolour' atau disingkat menjadi 'mfc'

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,7,2,6,2,8,4])


plt.plot(y,marker = 'o',ms=20,mfc='r') 
plt.show()
```

**Create labels for plot**

membuat label untuk plot dengan menggunakan keyword xlabel() dan ylabel()


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.xlabel("X Point")
plt.ylabel("Y point")

plt.plot(x,y,ms=20) 
plt.show()
```

**Create tittle for plot**

Berbeda dengan label, tittle merupakan judul dari diagram itu yang mana merupakan tema dari diagram tsb. Debgan menggunakan keyword tittle()



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'red','size':15}



plt.xlabel("X Point",fontdict = font2)
plt.ylabel("Y point",fontdict = font2)
plt.title("Percobaan data",fontdict=font1)

plt.plot(y,ms=20) 
plt.show()
```

**Size Dot of Scatter plot**

Mengubah ukuran dot dalam plot scatter, dapat menggunakan  argument s untuk mengubah function  size

sebagai contoh berikut : 

```python
import matplotlib.pyplot as plt
import numpy as np

#data harga car suzuki
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 260, 280, 290, 300, 310, 320, 330])

plt.scatter(x,y) 

#data harga car toyota
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([255, 260, 273, 220, 250, 290, 200, 310, 350, 330])

sizes = np.array([250,50,100,140,250,160,130,270,200,300])


plt.scatter(x,y,s = sizes)


plt.grid()
plt.show()
```

**Combine Size And Color of scatter plot**

Menggabungkan size dan color


```python
#dengan menggunakan cmap dan random randint


import matplotlib.pyplot as plt
import numpy as np

#data harga car suzuki
x = np.random.randint(100,size=(100))
y = np.random.randint(100,size=(100))
colors = np.random.randint(100,size=(100))

sizes = 10*np.random.randint(100,size=(100))

plt.scatter(x,y,s = sizes,c=colors,cmap='nipy_spectral',alpha=0.5)

plt.colorbar()
plt.show()
```

**Matplotlib Bars**

```python
import matplotlib.pyplot as plt
import numpy as np

#definition name of bars
x = np.array(["A","B","C","D","E"])
#definition array bars
y = np.array([4,5,2,8,7])

plt.suptitle("Data")
plt.bar(x,y)
plt.show()

```

**Matplotlib Histogram**

Histogram merupakan penyajian distribusi frekuensi dengan menggunakan diagram batang tegak. Histogram ini berkaitan dengan distribution data dan juga standar deviasi. untuk menggunakan histogram dalam matplotlib dapat memanggil function hist()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(150,30,270)

plt.suptitle("Data")
plt.hist(x,color = "yellow")
plt.show()
```

Labelling pie charts

```python
import matplotlib.pyplot as plt
import numpy as np

#create array of pie
x = np.array([35,50,41,75])
# create new variable to assignment of labels
xlabels = (["Jawa Barat","Jawa Tengah","Jawa Timur","Kalimantan"])

plt.pie(x,labels = xlabels) #declarate labels
plt.show()
```

**Explode Pie Charts**

Memecah atau memisahkan pie

```python
import matplotlib.pyplot as plt
import numpy as np

#create array of pie
x = np.array([35,50,41,75])
# create new variable to assignment of labels
xlabels = (["Jawa Barat","Jawa Tengah","Jawa Timur","Kalimantan"])

#exploding 
xExplode = [0,0,0.2,0]

plt.pie(x,labels = xlabels,explode = xExplode,shadow=True) #declarate explods 
plt.show()
```

Legend

```python
import matplotlib.pyplot as plt
import numpy as np

#create array of pie
x = np.array([35,50,41,75])
# create new variable to assignment of labels
xlabels = (["Jawa Barat","Jawa Tengah","Jawa Timur","Kalimantan"])

#exploding 
xExplode = [0,0,0.2,0]

#show or title
plt.pie(x,labels = xlabels,explode = xExplode,shadow=True) 
plt.legend(title = "Data persebaran BBM Subsidi")
plt.show()
```
